```python
from qiskit import QuantumCircuit, transpile
from qiskit_ibm_runtime import QiskitRuntimeService, Sampler
import numpy as np
import matplotlib.pyplot as plt
```

```python
# --- 设定实验参数 ---
qubit_index = 19           # 选择qubit编号（建议选T1接近median的）(424.73)
depths = list(range(0, 4000, 200))  # 0到4000，每500一个点
num_shots = 10000         # 每个depth的shots数

# --- 创建电路列表 ---
circuits = []
for depth in depths:
    qc = QuantumCircuit(1)
    qc.x(0)  # 准备 |1> 态
    for _ in range(depth):
        qc.id(0)  # 插入identity门
        qc.barrier()
    qc.measure_all()
    circuits.append(qc)

# --- 编译电路 (transpile) ---
transpiled_circuits = transpile(circuits, backend=backend, optimization_level=0)
```

```python
# --- 提交运行 ---
sampler = Sampler(backend)
job = sampler.run(transpiled_circuits, shots=num_shots)
result = job.result()

# --- 处理测量结果 ---
counts_list = [r.data.meas.get_counts() for r in result]
p1_list = []
for counts in counts_list:
    counts_1 = counts.get('1', 0)
    p1 = counts_1 / num_shots
    p1_list.append(p1)

# --- 绘制P(1) vs depth图 ---
plt.figure(figsize=(8,6))
plt.plot(depths, p1_list, 'o-', label='Experimental Data')
plt.xlabel('Number of Identity Gates')
plt.ylabel('P(1)')
plt.title(f'T1 Relaxation Measurement on Qubit {qubit_index}')
plt.grid(True)
```

```python
# Process results
shots=10000
decay_z = []
for r in result:
    dist = r.data.meas.get_counts()
    p0 = dist.get("0", 0)
    p1 = dist.get("1", 0)
    pz = (p0 - p1) / shots  # Expectation value of Z
    decay_z.append(pz)

# Plot
time_steps = [i * 200 for i in range(0, 20)]
plt.plot(time_steps, decay_z, 'o-', label='⟨Z⟩')
plt.xlabel("Number of identity gates (idling steps)")
plt.ylabel("⟨Z⟩ expectation value")
plt.title("Population decay under repeated identity gates")
plt.grid(True)
plt.show()
```
